Для первичного просмотра данных напишем функцию `quick_check`.

```python
# Функция для быстрого просмотра содержимого датафрейма и информации о нем.
def quick_check(dt, dist_cols=[], listing=1):
    """Функция выводит на экран общую информацию о датафрейме,
    первые строки датафрейма, распределение данных в указанных столбцах, 
    наиболее часто встречающиеся значения в указанных столбцах,
    а также черновые гистограммы
    """
    printmd('**Общая информация:**')  
    dt.info()
    
    printmd('**Первые строки таблицы:**')  
    display(dt.head())

    if dist_cols != []:
        printmd('**Распределение в столбцах:**')
        
    for column in dt.columns:
        if column in dist_cols:
            printmd("<span style='color:green'>**Распределение в столбце**</span>" + " " + dt[column].name + ":")
            
            if dt[column].nunique() > 2:
                display(dt[column].describe())
                dt[column].hist(bins=30)
                plt.show()
                
            printmd("Наиболее часто встречающиеся события в столбце" + " " + dt[column].name + ":")
            display(dt[column].value_counts(ascending=False).head())

    if listing == 1:
        display(dt.columns)
```

### Оценка признаков по группам

Посмотрим на внешний вид таблицы еще раз:

Разделив данные на две группы по факту оттока (0 - оттока не произошло, 1 - отток произошел), выведем на экран средние значения не-бинарных признаков. Для бинарных признаков посмотрим на процентное соотношение категорий по группам оттока.

Для этого напишем функцию `group_features_check`:

```python
# Определение функции:
def group_features_check(dt, feature):
    """
    Функция печатает процентное соотношение по группам признака feature
    для бинарных переменных, и печатает среднее значение столбца для всех других признаков.
    """
    for column in dt.columns:
        if dt[column].nunique() > 2:
            printmd("<span style='color:green'>**Среднее значение по группе в столбце**</span>" + " " + dt[column].name + ":")
            print(dt.groupby(feature)[column].mean())
            print()
        else:
            if dt[column].name == 'churn':
                print()
            else:
                all_count = dt.groupby(feature)[column].count()
                one_count = dt.groupby(feature)[column].sum()
                zero_count = all_count - one_count
                one_perc = one_count / all_count
                zero_perc = zero_count / all_count
            
                printmd("<span style='color:green'>**Доля значения 1 в столбце**</span>" + " " + dt[column].name + ":")
                print(one_perc)
                printmd("<span style='color:green'>**Доля значения 0 в столбце**</span>" + " " + dt[column].name + ":")
                print(zero_perc)
                print()
```

Применим функцию для просмотра результатов по группам:

```python
# Вызов функции:
group_features_check(gym_churn, 'churn')
```

### Гистограммы распределения признаков по факту оттока

Разделим данные на два датафрейма по признаку оттока:

```python
# Создание новых датафреймов по срезу данных.
gym_stay = gym_churn.loc[gym_churn['churn'] == 0]
gym_go = gym_churn.loc[gym_churn['churn'] == 1]
```

Построим гистограммы распределения признаков по группе оттока. Напишем функцию `show_hist`:

```python
def show_hist(df_1=gym_stay, df_2=gym_go, column="age", 
              name='Распределение возраста клиентов',
              xlabel='Возраст', ylabel='Количество клиентов'):
    """
    Функция строит 2 гисторгаммы на одном поле, 
    используя данные столбца с одинаковым именем 
    из двух разных датафреймов.
    Также выводит на экран среднее значение, медиану, 
    стандартное отклонение и дисперсию.
    """
    printmd("**Признак:**" + " " + xlabel)
    
    plt.figure(figsize=(10, 5))

    sns.set(style="darkgrid")

    sns.histplot(data=df_2, x=column, color="red", label="Ушли", kde=True)
    sns.histplot(data=df_1, x=column, color="skyblue", label="Остались", kde=True)

    plt.title(name, loc="center", fontsize=15)


    plt.ylabel(ylabel, fontsize=11);
    plt.xlabel(xlabel, fontsize=11);
    
    print('Показатели для клиентов, попавших в отток:')
    print('Среднее значение:', round(df_2[column].describe()[1], 2))
    print('Медиана:', round(df_2[column].describe()[5], 2))
    print('Стандартное отклонениее:', round(df_2[column].describe()[2], 2))
    #print('Дисперсия:', round(np.var(df_2[column], ddof=1), 4))
    print()
    print('Показатели для клиентов, оставшихся в фитнес-центре:')
    print('Среднее значение:', round(df_1[column].describe()[1], 2))
    print('Медиана:', round(df_1[column].describe()[5], 2))
    print('Стандартное отклонениее:', round(df_1[column].describe()[2], 2))
    #print('Дисперсия:', round(np.var(df_1[column], ddof=1), 4))
    
    plt.legend() 
    plt.show() 
```

#### Возраст

```python
# Применение функции для вывода распределения клиентов по возрасту.
show_hist()
```

Мы видим, что в среднем, клиенты, оставшиеся в фитнес-центре немного старше тех, кто попал в отток.

#### Дополнительные траты

```python
# Применение функции для вывода распределения дополнительных трат.
show_hist(column='avg_additional_charges_total',
          name='Распределение дополнительных трат',
          xlabel='Дополнительные траты, руб',
          ylabel='Количество клиентов'        
)
```

В среднем, дополнительные траты у клиентов, попавших в отток ниже, чем у клиентов, оставшихся в фитнес-центре.
```python
# Определение функции.
def plot_stacked(df, feature):
    """
    Функция строит столбчатые диаграммы распределения признака по группам оттока.
    """
    df = gym_churn.groupby(['churn', feature])[feature].count().unstack()
    df.plot(kind='bar', stacked=True, color=('#FF7F50','#40E0D0', '#7B68EE'), figsize=(4,6))
    
    plt.title(f'Распределение по признаку {feature} в группах')
    plt.xticks(rotation=0, ha='center')
    plt.xlabel('Отток')
    plt.ylabel('Количество пользователей')
    plt.show();
```

## Прогнозирование оттока клиентов

### Построение модели логистической регрессии

Зададим переменные для модели. Уберем из данных коррелирующие признаки:

```python
X = gym_churn.drop(['churn','avg_class_frequency_current_month', 'contract_period'], axis = 1)
y = gym_churn['churn']
```

Разделим данные на обучающую и валидационную(тестовую) выборку:

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

Обозначим функцию вывода на экран основных метрик:

```python
# Определение функции по выводу метрик на ээкран.
def print_all_metrics(y_true, y_pred, y_proba, title='Метрики классификации'):
    print(title)
    print('\tAccuracy: {:.2f}'.format(accuracy_score(y_true, y_pred)))
    print('\tPrecision: {:.2f}'.format(precision_score(y_true, y_pred)))
    print('\tRecall: {:.2f}'.format(recall_score(y_true, y_pred)))
    print('\tF1: {:.2f}'.format(f1_score(y_true, y_pred)))
    print('\tROC_AUC: {:.2f}'.format(roc_auc_score(y_true, y_proba)))
```

Обучим модель:

```python
# Обучение модели на обучающей выборке.
scaler = StandardScaler()
scaler.fit(X_train)

# Преобразование в обучающий и валидационные наборы данных7
X_train_st = scaler.transform(X_train)
X_test_st = scaler.transform(X_test)
```

Построим модель и выведем ее метрики:

```python
# Алгоритм для модели логистической регрессии.
lr_model = LogisticRegression(random_state=0, solver='liblinear')

# Обучение модели.
lr_model.fit(X_train_st, y_train)

# Прогнозирование на основе обученной модели.
lr_predictions = lr_model.predict(X_test_st)
lr_probabilities = lr_model.predict_proba(X_test_st)[:, 1]

# Вывод всех метрик на экран.
print_all_metrics(
    y_test,
    lr_predictions,
    lr_probabilities,
    title='Метрики для модели логистической регрессии:',
)
```

Метрики выглядят неплохо. Особенно хороша метрика ROC AUC, которая отражает общее качество модели.

Мы знаем, что клиентов, попавших в отток в 3 раза меньше, чем клиентов, оставшихся в фитнес-центре. Посмотрим, меняет ли метрики модели изменение порога:

```python
threshold = 0.25

# на основании вероятностей и соотношения классов рассчитайте predict
custom_predictions = [0 if i < threshold else 1 for i in lr_probabilities]

# выведите все изученные метрики для прогноза по новому порогу
print('Метрики для прогноза с кастомным порогом')
print(
    'Accuracy for custom: {:.2f}'.format(
        accuracy_score(y_test, custom_predictions)
    )
)
print(
    'Precision for custom: {:.2f}'.format(
        precision_score(y_test, custom_predictions)
    )
)
print(
    'Recall for custom: {:.2f}'.format(
        recall_score(y_test, custom_predictions)
    )
)
print('F1 for custom: {:.2f}'.format(f1_score(y_test, custom_predictions)))
```

С заданным порогом показатели модели ухудшились, доля правильных прогнозов - всего 85%. Точность значительно снизилась, но полнота на хорошем уровне.

**Рекомендации по удержанию клиентов:**

- Стимулировать более частые посещения фитнес-центра, в том числе дополнительные услуги - чем больше времени клиент проводит в центре, тем лучше.


- Первый месяц в фитнес-центре важен для удержания клиента - есть смысл помочь ему сформировать привычку посещения фитнес-центра. Возможно, в формате рассылки для новичков, рассказать, как полезны для здоровья и настроения регулярные занятия спортом, что занятия не должны быть в тягость, что каждый клиент может выбрать то, что ему по душе. Рассказать реальные истории клиентов - как совмещать спорт, работу, семью и отдых.



- Акцент на групповые занятия. Социализация и нахождение в группе единомышленников (в данном случае - единодеятельников) может повышать уровень положительных эмоций и заставлять клиентов приходить чаще.


- Регулярно проводить акцию 'приведи друга'. Возможно, стоит предлагать клиентам, которые попадают в склонные к оттоку кластеры, приводить друзей для их собственного удержания. 


- Клиенты, пришедшие в фитнес-центр по партнерским программам, лучше удерживаются, то есть есть смысл расширять список партнеров.



- Стимулировать заключение долгосрочного договора.



- Подключать дополнительные усилия по удержанию клиента за 2 месяца до истечения контракта - специальные акции.


- Если средняя посещаемость клиента падает более, чем на 20%, отправлять ему спец предложение, например "Класс по водной гимнастике и шоколадный торт в кафетерии в подарок".
